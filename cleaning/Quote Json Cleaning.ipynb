```python
import pandas as pd
import requests
import time
```

```python
quotedf = pd.read_json('quotes.json')
```

### Shuffling

```python
quotedf.drop_duplicates(subset ="Quote", keep = "first", inplace = True)
quotedf.sort_values("Popularity", inplace = True, ascending=False)
quotedf = quotedf[:800]
#quotedf = quotedf[['Quote', 'Author']]
quotedf = quotedf.sample(frac=1)
quotedf
```

### Fixing Author

```python
quotedf = pd.concat([quotedf, quotedf['Author'].str.split(', ', expand=True)], axis=1)
quotedf = quotedf.drop(['Author', 1,2], axis=1)
quotedf = quotedf.rename({0: 'Author'}, axis=1)
quotedf
```

### Check DDG for data

```python
authors = quotedf['Author'].values
```

```python
summaries = []
images = []
count = 0
for author in authors:
    ddgjson = requests.get("https://api.duckduckgo.com/?q=" + author + "&format=json").json()
    summaries.append(ddgjson['AbstractText'])
    images.append(ddgjson['Image'])
    count = count + 1
    if (count == 100):
        time.sleep(3)
        count = 0
```

```python
modquotedf = quotedf[:752]
modquotedf['Summary'] = summaries
modquotedf['Image'] = images
modquotedf = modquotedf[modquotedf['Image'] != '']
modquotedf
```

```python
modquotedf.to_json('491quotes.json', orient="records")
```
